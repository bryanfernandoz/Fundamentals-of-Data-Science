```python
state=pd.read_csv("/content/US_violent_crime.csv")
state.head()
```

```python
def some_func(x):
  return x*2
state.apply(some_func)
state.apply(lambda n: n*2)
```

```python
state.transform(func = lambda x : x * 10)
```
